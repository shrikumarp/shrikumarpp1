```python
from sklearn.datasets import load_boston
```

```python
boston= load_boston()
```

```python
print(boston.DESCR)
```

```python
bostonhousng= boston.data
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
print(boston['feature_names'])
```

```python
bostonhousng_df= pd.DataFrame(bostonhousng, columns= boston['feature_names'])
```

```python
target_df= pd.DataFrame(boston['target'], columns= ['MEDV'])
```

```python
sns.distplot(target_df)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(bostonhousng_df, target_df, test_size=0.4, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm= LinearRegression()
```

```python
lm.fit(X_train, y_train)
```

```python
print(lm.intercept_)
```

```python
lm.coef_
```

```python
predictions = lm.predict(X_test)
```

```python
plt.scatter(y_test,predictions)
```

```python
sns.distplot((y_test-predictions),bins=50);
```

The plot between the y_test values and predictions is linearly correlated, and the difference/error distribution is normal, which means that the model is a good fit.


```python
from sklearn import metrics
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```
